### Django

1. 파이썬 기반의 무료 오픈소스 웹애플리케이션 프레임워크(Open Source Web Application Framework)
 - 장고웹프로그래밍의 특징은 개발하는데 쉽고 빠르다. 이는 장고의 개발방식이 MTV(Model, Template, View)에 따른 일정한 룰에 의해 진행되고 웹프로그래밍에서 공통적으로 필요한 기능들을 사전에 만들어 둔 후에 단축함수, 제네릭뷰등으로 제공하기 때문아디.

  - 빠른 개발속도, 개발비용 절감
  - 코드 완성도를 높게 유지할 수 있으며 확장성이 좋다.
  - 사용자인증, 사용자관리등의 기능이 기본적으로 구현되어 있다.
 - 성공적인 도입사례는 `instagram`이다.

3. MTV개발방식

<img src="../images/01.django설치_01.png" width="500" height="400"/>

>[MVC와 MTV 비교]
* MVC        MTV      설명
* ==============================================================================
* Model      Model    데이터베이스와 관련된 처리를 담당하는 코드
* View       Template 사용자가 보게되는 화면을 정의하는 코드
* Controller View     데이터를 처리한 결과를 템플릿에 전달하는 코드(컨트롤러역할)

장고의 MTV방식은 자바 웹프로그램의 MVC방식과 거의 동일한 개념으로 웹프로그래밍영역을
3가지개념으로 나누어서 개발하는 방식이다.

테이블을 정의하는 Model, 사용자 보게될 화면을 정의하는 Template, 애플리케이션의 제어흐름 및
처리 로직을 정의하는 View로 구분해서 개발을 진행한다.

장고에서 프로젝트를 생성하기 위해 startproject 및 startapp명령을 실행하면 자동으로 프로젝트의 뼈대에 해당하는 디렉토리와 파일을 생성한다.

모델은 `models.py`파일에 템플릿은 templates디레토리하위dml `*.html`파일에 뷰는 `views.py`파일에 작성하도록 기본골결을 만들어 준다.

1) MTV 코딩순서

   (1) 프로젝트뼈대 만들기 : 프로젝트 및 앱개발에 필요한 디렉터리와 파일 생성
   (2) 모델코딩하기 : 테이블과 관련하상을 개발(models.py, admin.py파일)
   (3) URLconf 코딩하기 : URL 및 뷰 매핑 관계를 정의(urls.py파일)
   (4) 뷰코딩하기 : 애블리케이션 로직개발(views.py파일)
   (5) 템플릿코딩하기 : 화면 UI개발(templates/*.html파일)

2) settings.py

   이 파일은 프로젝트의 설정파일이다. 프로젝트를 생성하면 장고가 기본사항들을 자동으로 
   생성해주고 추가적으로 필요한 사항은 개발자가 원하는 내용을 등록한다.

   <주요사항>
   (1) 데이터베이스 설정 : 기본으로 sqlite3 데이터베이스 엔진을 사용하는 것을 지정
   (2) 템플릿항목 설정 : TEMPLATES항목을 지정
   (3) 정적파일항목 설정 : STATIC_URL등 관련항목을 설정
   (4) 애플리케이션 등록 : 개발자가 개발하는 앱. 즉, 프로젝트에 포함되는 애플리케이션을 등록
   (5) 타임존을 설정 : 최초에는 세계표준시(UTC)로 설정되어 있는 것을 한국시간으로 변경       

3) models.py

   테이블을 정의하는 파일이다. 장고의 특징중 하나로 데이터베이스처리는 ORM(Object Relation
   Mapping)기법을 사용한다. 즉, 테이블을 클래스로 매핑해서 테이블에 대한 CRUD기능을 클래스
   객체에 대해 수행하면 장고가 내부적으로 테이터베이스에 매핑해 주는 방식이다.

   장고에서는 ORM기법에 따라 테이블 하나의 클래스로 정의하고 테이블의 컬럼은 클래스의 변수
   (속성)로 매핑한다. 테이블클래스는 django.db.models.Model클래스를 상속받아 정의하고 각
   클래스 변수 타입도 장고에서 미리 정의해 놓은 필드 클래스를 사용한다.

   테이블의 신규생성, 변경등 models.py파일에서 데이터베이시 변경사항이 발생하면 데이터베이스에
   변경사항을 반영해주는 작업을 해야 하는데 이런 작업을 마이그레이션(migration)을 해줘야 한다.

   마이그레이션이란 테이블 및 필드의 생성,삭제,변경 등과 같이 데이터베이스에 대한 변경사항을
   전달하는 정보이다. 

   물리적으로는 애플리케이션 디렉토리별로 migrations/디렉토리하위에 마이그레이션 파일들이 존재
   한다. 장고는 이런 마이그레이션 정보를 참조해서 변경사항을 실제 데이터베이스에 반연하는
   makemigrations 및 migrate명령을 제공한다.       


4) URLconf

   URL과 view(함수 또는 메서드)를 매핑해 주는 urls.py파일을 말한다. URLconf를 정의할 때는
   하나의 파일에 또는 두개의 파일에 정의할 수 있다.

   보통 두번쨰 방법을 추천. 즉, 프로젝트전체URL을 정의하는 프로젝트URL과 앱마다 정의하는 
   앱URL 2개층으로 나누어서 코딩하는 방법이다.

   이 방식은 URLconf모듈을 계층적으로 구성하기 때문에 변경 또는 확장이 용이한다. 
   URL패턴별로 이름을 지정할 수 있고 패턴그룹에 대해 namespace를 지정할 수도 있다. 이는
   reverse()함수나 {% url %} 탬플릿 태그를 사용해서 URL을 하드코딩하지 않고 필요한 URL을
   추출할 수 있도록 한다.

5) views.py

   views.py는 뷰 로직을 코딩하는 가장 중요한 파일이다. 간단한 로직이라면 몇줄만 코딩하면
   되지만 프로젝트 개발이 커지고 로직이 복잡해 질 경우 당연히 views.py파일도 복잡해지고
   코딩량도 많아진다. 따라서 가독성과 유지보수, 재활용등을 고려해야 된다.

   장고에서는 뷰로직을 함수로 코딩할 것인지 클래스로 코딩할 것인지에 따라 함수형뷰(function
   based vew)와 클래스형뷰(class based view)로 구분한다.

   장고는 다른 프레임워크에 비해 뷰작성에 편리한 기능을 많이 제공하고 있다. 대표적인 것이
   단축함수 및 클래스형 제네릭 뷰를 제공한다.

6) template

   웹화면(웹페이지)별로 템플릿파일(*.html)이 하나씩 필요하므로 개발시에 여러개의 템플릿파일을
   작성하게 되고 이런 파일을 한 곳에 모아두기위해 템플릿디렉토리가 필요한다.

7) Admin 사이트

   Admin사이트는 테이블의 내용을 열람하고 수정하는 기능등을 제공한다. 이 사이트체서 User와
   Group테이블을 포함해, 개발자가 만드는 테이블에 대한 데이터의 입력, 수정, 삭제등의 작업을
   할 수가 있다. 

   Admin화면에 기본적을 User와 Group테이블이 보이는 것은 settings.py파일에 django.conrtib.auth
   애플리케이션이 등록되어 있기 때문이다.

   Admin기능은 자주 사용되는 편리한 기능으로 필요할 때 언제든지 sql없이도 테이블의 내용확인,
   테이블의 수정,삭제를 할 수있다.

   Admin 사이트에 원하는 테이블을 등록하기 위해서는 admin.py파일에 작업하면 된다.


8) 개발용웹서버 - runserver

   장고는 개발과정에서 필효한 현재 웹프로그램을 실행할 수 있도록 runserver라는 테스트용 웹서버를
   제공한다.

   하지만 실제 고객에게 오픈하는 상용화를 고려한다면 Apache또는 Nginx등의 웹서버를 사용해야한다.
   runserver는 상용 웹서버에 비해 처리능력과 보안에 취약하기 때문이다.
